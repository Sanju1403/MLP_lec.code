```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
#reading heart disease classification dataset.
df = pd.read_csv("/content/drive/MyDrive/Data Sets/mlp-main/data/heart_statlog_cleveland_hungary_final.csv")
df
```

```python
#extracting input and output data from the training dataset
#excluding all column, except a last one

x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
#1190 samples and 11 features
print(x.shape)
#binary targets
print(y.shape)
```

```python
#splitting dataset into train and test sets

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 24, stratify=y)
x_train
```

```python
#scaling the features

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_test
```

```python
#input layer: 11 input features
#hidden layer1: 16 nodes (using ReLU activation function)
#output layer: 2 classes (using sigmoid because it's a classification problem)

#defining the MLP model

mlp_cl = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation='relu', input_shape = (11,)),
    tf.keras.layers.Dense(16, activation = 'relu'),
    tf.keras.layers.Dense(1, activation = 'sigmoid')
])
```

```python
mlp_cl.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
# train the model

mlp_cl.fit(x_train, y_train, epochs = 100, validation_split = 0.2)
```

```python
#evaluating the model (after training, evaluate the mlp_cl model on the test set (X_test, y_test) to see how well the model is performing)

loss, accuracy = mlp_cl.evaluate(x_test, y_test)
print(f'Test accuracy: {accuracy:.4f}')
```

```python
# Making prediction

prediction = mlp_cl.predict(x_test)
predicted_classes = (prediction > 0.5).astype(int)
print(f'Predicted classes', predicted_classes.flatten())
```
